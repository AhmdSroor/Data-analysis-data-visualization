```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
data = pd.read_csv('Top 5 Leagues_ Average Expected Goal Contributions (2014-2020) - _u_pastenague - Top 5 Leagues_ xG Contributions (2014-2020).csv')
```

```python
data
```

```python
data.info()
```

```python
#Start from highset xGA_ratio
data = data.sort_values(by='xGA_ratio', ascending=False)
```

```python
data
```

```python
#Filter data by these conditions to get reasonable patch
filter1 = data.matches > 25
filter2 = data.mins > 2000
filter3 = data.xGA_ratio >= 0.1
```

```python
data = data[filter1 & filter2 & filter3]
```

```python
data
```

```python
#Top 50 includes only 2 Premier League players
data.head(50)
```

```python
#Has no meaning but only the top of 2016-17 season (the premier league player)

vis = sns.lmplot( data = data, x = 'season', y = 'xGA_ratio', fit_reg = False, hue = 'league', size = 10  )
```

```python
#We can split by League
Bundesliga = data[data.league == 'Bundesliga']
La_Liga = data[data.league == 'La Liga']
Serie_A = data[data.league == 'Serie A']
Premier_League = data[data.league == 'Premier League']
Ligue_1 = data[data.league == 'Ligue 1']

#Or by Season

season1415 = data[data.season == '2014-15']
season1516 = data[data.season == '2015-16']
season1617 = data[data.season == '2016-17']
season1718 = data[data.season == '2017-18']
season1819 = data[data.season == '2018-19']
season1920 = data[data.season == '2019-20']
```

```python
Bundesliga
```

```python
Bundesliga.head(20)
```

```python
sns.set(style = "darkgrid", palette = "muted", color_codes = True)

ax = sns.boxplot(data = Bundesliga, x = 'season', y = 'xGA_ratio', orient = 'v', color = 'lightgray', showfliers = False)

plt.setp(ax.artists, alpha=0.5)

sns.stripplot(x = 'season', y = 'xGA_ratio', data = Bundesliga, jitter = True, size = 8, linewidth = 0, hue = 'player', alpha = 0.7)

ax.axes.set_title('Bundesliga Stats',fontsize=30)
ax.set_xlabel('xGA_ratio',fontsize=20)
ax.set_ylabel('Season',fontsize=20)

# Define where to place the legend
ax.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```
